```python
forecast_out = 45
amzn_prc['Prediction'] = amzn_prc['Adj Close'].shift(-forecast_out)
print(amzn_prc.tail())
```

```python
X = np.array(amzn_prc.drop(['Prediction'], 1))
X = X[:-forecast_out]
```

```python
y = np.array(amzn_prc['Prediction'])
y = y[:-forecast_out]
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr_confidence = lr.score(x_test, y_test)
print("lr confidence: ", lr_confidence)
```
